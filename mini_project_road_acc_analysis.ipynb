```python
from sklearn.preprocessing import StandardScaler
```

```python
x = road_accident_analysis.iloc[:,1:]
x.head()
```

```python
sc = StandardScaler()
x_sc = sc.fit_transform(x)
print('Standardized data\n',x_sc)
```

```python
print(type(x_sc))
```

```python
print(x_sc.shape)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
x_res = pca.fit_transform(x_sc)
print(x_res.shape)
# x_res = principal_components
x_res
```

```python
eig_val = pca.explained_variance_        # eig_val
eig_vec = pca.components_                # eig_vec
print(eig_val)
print(eig_vec)
```

```python
tot = sum(eig_val)
var_eig_val = [(i/tot)*100 for i in eig_val]
print(var_eig_val)
print(sum(var_eig_val))
cum_var_eig_val = np.array(var_eig_val).cumsum()
print(cum_var_eig_val)
```

```python
plt.pie(x=var_eig_val,labels=['EVec1','EVec2','EVec3','EVec4','EVec5'],
       autopct='%.2f%%')
plt.title('Variance retained  by different Eigen values')
plt.show()
```

```python
print(x_sc.shape)
print(eig_vec.shape)
```

```python
pc_res  = np.dot(x_sc,eig_vec.T)
print(pc_res)
```

k-means

```python
from sklearn.cluster import KMeans
```

```python
# wcss = within cluster sum of square
wcss = []
for i in range(1,12): # i = no of clusters,
    m = KMeans(n_clusters=i)
    m.fit(x)
    wcss.append(m.inertia_)

print('WCSS\n',wcss)
plt.plot(list(range(1,12)),wcss)
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
for i in range(2,12): # i = no of clusters
    m = KMeans(n_clusters=i).fit(x)
    print(f'Silhoutte Score at K={i}',silhouette_score(x, m.labels_))
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
m2 = KMeans(n_clusters=2)
visualizer = SilhouetteVisualizer(m2, colors='yellowbrick')
visualizer.fit(x) 
plt.show()
```

```python
k2 = KMeans(n_clusters=1)
k2.fit(x)
ypred2 = k2.predict(x)
print(ypred2)
```

```python
cent2 = k2.cluster_centers_
print(cent2)
```
